1 EXTRACT

```python
from google.cloud import bigquery

# Replace with your actual project ID
client = bigquery.Client(project="project-cf3503bf-b7cb-4c6e-bcc")
```

```python

import pandas as pd


# SQL query
query = """
SELECT 
  c.Company_Name,
  i.Industry_Name,
  h.Headquarters_Name,
  f.Rank,
  f.Sales_Billion,
  f.Profit_Billion,
  f.Assets_Billion,
  f.MarketValue_Billion
FROM `project-cf3503bf-b7cb-4c6e-bcc.forbesdataset.fact_financials` f
JOIN `project-cf3503bf-b7cb-4c6e-bcc.forbesdataset.dim_company` c 
  ON f.Company_ID = c.Company_ID
JOIN `project-cf3503bf-b7cb-4c6e-bcc.forbesdataset.dim_industry` i 
  ON c.Industry_ID = i.Industry_ID
JOIN `project-cf3503bf-b7cb-4c6e-bcc.forbesdataset.dim_headquarters` h 
  ON c.HQ_ID = h.HQ_ID
"""

# Extract data into DataFrame
df_raw = client.query(query).to_dataframe()
print(df_raw.head())
```

2 . Transform (clean & enrich)
Add calculated fields like Profit Margin and Efficiency Ratios:

```python
# Handle missing values
df_raw = df_raw.dropna()

# Create derived metrics
df_raw["Profit_Margin_%"] = (df_raw["Profit_Billion"] / df_raw["Sales_Billion"]) * 100
df_raw["MarketValue_to_Assets"] = df_raw["MarketValue_Billion"] / df_raw["Assets_Billion"]

# Normalize industry names (example)
df_raw["Industry_Name"] = df_raw["Industry_Name"].str.strip().str.title()

print(df_raw.head())
```

3. Load (to Power BI or another destination)

```python
# Save to CSV for Power BI import
df_raw.to_csv("global_business_clean.csv", index=False)
```

Or push back into BigQuery:

```python
# Define destination table
table_id = "project-cf3503bf-b7cb-4c6e-bcc.forbesdataset.cleaned_financials"

job = client.load_table_from_dataframe(df_raw, table_id)
job.result()  # Wait for job to complete
print("Data loaded successfully into BigQuery!")
```

What This Gives You
Extract: Pull raw data from BigQuery.

Transform: Clean, enrich, and calculate business metrics.

Load: Save for Power BI dashboards or reload into BigQuery for reuse.

This pipeline is reusable â€” you can rerun it whenever Kaggle updates the dataset or you want fresh insights.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set style
sns.set(style="whitegrid")

# 1. Bar chart: Top 10 companies by Sales_Billion
top_sales = df_raw.nlargest(10, "Sales_Billion")
plt.figure(figsize=(12,6))
sns.barplot(x="Sales_Billion", y="Company_Name", data=top_sales, palette="Blues_r")
plt.title("Top 10 Companies by Sales (Billion USD)")
plt.xlabel("Sales (Billion USD)")
plt.ylabel("Company")
plt.tight_layout()
plt.savefig("top10_sales.png", dpi=300)
plt.show()

# 2. Scatter plot: Sales vs Profit, colored by Industry
plt.figure(figsize=(10,6))
sns.scatterplot(
    x="Sales_Billion", 
    y="Profit_Billion", 
    hue="Industry_Name", 
    data=df_raw, 
    alpha=0.7
)
plt.title("Sales vs Profit by Industry")
plt.xlabel("Sales (Billion USD)")
plt.ylabel("Profit (Billion USD)")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.savefig("sales_vs_profit.png", dpi=300)
plt.show()

# 3. Bubble chart: Market Value vs Assets, bubble size = Profit
plt.figure(figsize=(10,6))
sns.scatterplot(
    x="Assets_Billion", 
    y="MarketValue_Billion", 
    size="Profit_Billion", 
    hue="Industry_Name", 
    data=df_raw, 
    alpha=0.6, 
    sizes=(50, 500)
)
plt.title("Market Value vs Assets (Bubble size = Profit)")
plt.xlabel("Assets (Billion USD)")
plt.ylabel("Market Value (Billion USD)")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.savefig("marketvalue_vs_assets.png", dpi=300)
plt.show()

# 4. Boxplot: Profit Margin % distribution by Industry
plt.figure(figsize=(12,6))
sns.boxplot(x="Industry_Name", y="Profit_Margin_%", data=df_raw, palette="Set3")
plt.title("Profit Margin % Distribution by Industry")
plt.xlabel("Industry")
plt.ylabel("Profit Margin (%)")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.savefig("profit_margin_distribution.png", dpi=300)
plt.show()
```

```python
import pandas as pd

# FACT TABLE: Financials
fact_financials = df_raw[[
    "Company_Name", "Rank", "Sales_Billion", 
    "Profit_Billion", "Assets_Billion", "MarketValue_Billion",
    "Profit_Margin_%", "MarketValue_to_Assets"
]]

# DIMENSION TABLE: Company (links to Industry + HQ)
dim_company = df_raw[[
    "Company_Name", "Industry_Name", "Headquarters_Name"
]].drop_duplicates()

# DIMENSION TABLE: Industry
dim_industry = df_raw[["Industry_Name"]].drop_duplicates()

# DIMENSION TABLE: Headquarters
dim_headquarters = df_raw[["Headquarters_Name"]].drop_duplicates()

# Preview
print("Fact Financials:\n", fact_financials.head())
print("Dim Company:\n", dim_company.head())
print("Dim Industry:\n", dim_industry.head())
print("Dim Headquarters:\n", dim_headquarters.head())
```

```python
fact_financials.to_csv("fact_financials.csv", index=False)
dim_company.to_csv("dim_company.csv", index=False)
dim_industry.to_csv("dim_industry.csv", index=False)
dim_headquarters.to_csv("dim_headquarters.csv", index=False)
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a directed graph
G = nx.DiGraph()

# Add nodes
G.add_node("Fact_Financials")
G.add_node("Dim_Company")
G.add_node("Dim_Industry")
G.add_node("Dim_Headquarters")

# Add edges (relationships)
G.add_edge("Fact_Financials", "Dim_Company", label="Company_Name")
G.add_edge("Dim_Company", "Dim_Industry", label="Industry_Name")
G.add_edge("Dim_Company", "Dim_Headquarters", label="Headquarters_Name")

# Position nodes in a layout
pos = nx.spring_layout(G, seed=42)

# Draw nodes
nx.draw_networkx_nodes(G, pos, node_size=3000, node_color="skyblue")

# Draw edges
nx.draw_networkx_edges(G, pos, arrowstyle="->", arrowsize=20)

# Draw labels
nx.draw_networkx_labels(G, pos, font_size=10, font_weight="bold")

# Draw edge labels
edge_labels = nx.get_edge_attributes(G, "label")
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_size=9)

# Show plot
plt.title("Star Schema: Fact & Dimension Tables", fontsize=14)
plt.axis("off")
plt.show()
```
